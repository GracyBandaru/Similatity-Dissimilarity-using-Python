# pearson correlation

```python
import numpy as np
import matplotlib.pyplot as plt
experience=np.arange(5,15)
salary=np.array([10,15,16,17,18,20,25,26,29,30])
np.corrcoef(experience,salary)
```

```python
plt.scatter(experience,salary)
plt.xlabel('experience')
plt.ylabel('salary')
plt.show()
```

# cosine similarity

```python
import numpy as np
from numpy.linalg import norm
A = np.array([1,1,1,1,1,1,0,0,0])
B= np.array([0,0,1,0,1,1,1,1,1])
print("A:", A)
print("B:", B)
cosine = np.dot(A,B)/(norm(A)*norm(B))
print("Cosine Similarity:", cosine)
```

# Jaccard similarity

```python
import numpy as np
def jaccard_binary(x,y):
    intersection = np.logical_and(x, y)
    union = np.logical_or(x, y)
    similarity = intersection.sum() / float(union.sum())
    return similarity
x = [0,1,0,0,0,1,0,0,1]
y = [0,0,1,0,0,0,0,0,1]
simxy = jaccard_binary(x,y)


print(' Similarity between x and y is', simxy)
```

# Euclidean distance
